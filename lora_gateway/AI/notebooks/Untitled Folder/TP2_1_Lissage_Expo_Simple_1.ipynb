# Devoir INF312 (Lissage exponentiel simple)

### Note : On se limitera à $50$ valeurs et à un base de prévision $n = 30$

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
n = 30
alp = [0.001, 0.005, 0.01, 0.05, 0.1, 0.3, 0.5, 0.9]
```

```python
class TimeLissage:
    def __init__(self, file_path : str):
        with open(file_path, 'r') as f:
            cols = f.readline().strip().split(',') 
            
        self.cols = cols
        data = pd.read_csv(file_path)
        print(data.head())
        self.index = 1
        self.x = np.array([i for i in data[self.cols[self.index]]])
        self.x = self.x[:50]
        
    def lissage(self, alpha, n : int):
        x_chap = np.zeros(self.x.shape[0])
        x_chap[0] = self.x[0]
        tmp = 0
        for i in range(1, self.x.shape[0]):
            if i == n:
                for j in range(n-1):
                    tmp += (alpha ** j) * self.x[n-j]
                x_chap[n:] = (1-alpha) * tmp
                break
            x_chap[i] = x_chap[i-1] + (1-alpha)*(self.x[i]-x_chap[i-1])
        return x_chap
    
    def graphe(self):
        plt.figure(figsize=(20, 10))
        k = 1
        for i in alp:
            plt.subplot(2, 4, k)
            plt.plot(self.x, label=f'Série {self.cols[self.index]}')
            plt.plot(self.lissage(i, n), label=f'alpha = {i}', ls='--')
            plt.axvline(n, c='black', linewidth=0.9)
            plt.legend()
            k += 1
        plt.subplots_adjust(wspace=0.4, hspace=0.4)
        plt.legend()
        plt.show()

    def erreur(self):
        list_err = []
        for i in alp:
            err = np.square((self.x[:n] - self.lissage(i, n)[:n])).mean()
            print(f'Alpha : {i} -- Erreur : {err}')
            list_err.append(err)
```

```python
beer = TimeLissage('./dataset/monthly-beer-production-in-austr.csv')
beer.graphe()
beer.erreur()
```
